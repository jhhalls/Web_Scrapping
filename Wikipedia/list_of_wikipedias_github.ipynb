## extracting Table 2

```python
import re
# as per recommendation from @freylis, compile once only
CLEANR = re.compile('<.*?>') 

def cleanhtml(raw_html):
    cleantext = re.sub(CLEANR, ' \n', raw_html)
    return cleantext
```

```python
table_2_raw = soup.find_all('table')[1]
```

```python
table_2_list_of_lists = []
for i in table_2_raw.find_all('tr'):
    table_2_list_of_lists.append(cleanhtml(str(i)).replace("\n", '').split('   '))
```

```python
table_2_list_of_lists
```

```python
# removing all the extra spaces from inner lists
table_2_final_list = []
for i in table_2_list_of_lists:
    j = [k.strip() for k in i if k]    # removing the trailing spaces
    n = [y for y in j if y]      #removing extra space elements from the inside list
    table_2_final_list.append(n)
```

```python
table_2_final_list
```

```python
delete = ['(locked)', '(closed)', '(deleted)', '( deleted )', '( Taraškievica )', '( Kurmanji )' ]

for i in table_2_final_list:
    for j in i:
        if j in delete:
            i.remove(j)        
```

```python
[i for i in table_2_final_list if len(i)>5]
```

### Column names

```python
# column 1
table_2_column_1 = ' '.join(table_2_final_list[0][0].split()[:2])
# column 2
table_2_column_2 = ' '.join(table_2_final_list[0][0].split()[2:3])
# column 3
table_2_column_3 = ' '.join(table_2_final_list[0][0].split()[3:4])
# column 4
table_2_column_4 =  ' '.join(table_2_final_list[0][0].split()[4:6])
# column 5
table_2_column_5 = ' '.join(table_2_final_list[0][0].split()[6:8])
```

```python
table_2_column_names = [table_2_column_1, table_2_column_2, table_2_column_3, table_2_column_4, table_2_column_5]
```

```python
for i in table_2_column_names:
    print(i)
```
